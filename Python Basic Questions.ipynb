**9) What is type casting in Python ?**

Type casting or type conversion in Python is the process of converting a variable from one data type to another.

1) Implicit type casting - Python automatically converts one data type into another without losing data.

2) Explicit type casting - In explicit type casting, we manually convert one type to another using built in functions.

**12) How does the elif statement work ?**

The elif (else if) statement in Python is used for multiple condition checking. It allows the program to check multiple conditions one by one and execute the first one that evaluates to True.

```python
temperature = 30

if temperature > 35:
    print("It's too hot!")
elif temperature > 25:
    print("The weather is warm.")
elif temperature > 15:
    print("It's a bit chilly.")
else:
    print("It's cold.")
```

**13) What is the difference between for and while loops ?**

Loops in Python are used for repeating a block of code multiple times. The two main loops are:

- for loop – Used when the number of iterations is known.

- while loop – Used when the number of iterations is unknown, and it runs until a condition is False.
